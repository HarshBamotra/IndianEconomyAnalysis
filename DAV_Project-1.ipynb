## Data cleaning and pre-processing 
-----------------------------------------------
### Handling missing values

*Checking for missing or Nan values*

```python
sgdp.isna().sum()
```

*Missing value percentage in the dataset*

```python
sgdp.isna().sum()/len(sgdp)*100
```

*Dropping attribute with 25% missing or Nan values from the dataset*

```python
# total of 11 rows so each state must contain more than 25% of the data for furthur details
half = len(sgdp) / 4
x = sgdp.isna().sum() > half
drop_indexes = sgdp.columns[x]
# dropping columns from the dataframe 
sgdp.drop(drop_indexes, axis=1, inplace=True)
sgdp.shape
```

### Data Transformation

*Transposing the dataframe*

```python
sgdp = sgdp.set_index("State/Union Territory", drop = True)
```

```python
sgdp = sgdp.transpose()
sgdp.index.name = "Year"
sgdp
```

*Checking for duplicate data*

```python
sgdp.duplicated()
```

## Data analysis and Visulisation

```python
percent_change = sgdp.pct_change()
percent_change = percent_change*(100)
percent_change
```

```python
percent_change.mean().round(2).sort_values().plot.bar(title = "% Growth of each state", 
                                                      xlabel = "States", 
                                                      ylabel = "% change")
```

*Compairing the GDP growth rates of my home state(Uttarakhand, Himanchal Pradesh) with that of INDIA*

```python
#calulating all india GDP
sgdp["India"] = sgdp.sum(axis = 1)  
sgdp.head()
```

```python
perChange = sgdp.pct_change()*100
home_state = perChange[["Himachal Pradesh", 'Uttarakhand', 'India']]
comapred_states = home_state.sum()/ home_state.count()
comapred_states.plot(kind='bar',stacked=True, colormap = 'Dark2')
plt.title("Avg. % Growth of Home State vs National Avg. for Duration 2011-20", fontweight = 'bold')
plt.xlabel("States", fontweight = 'bold')
plt.ylabel("% growth rate", fontweight = 'bold')
```

*Plotting the GDP of all states of last year (2020- 2021)*

```python
result = sgdp.loc["2019-20"].copy()
result.drop(labels='India', inplace=True)
result.sort_values(inplace=True)
result.plot(kind='bar',stacked=True, title = "State Wise GDP in 2019-20", ylabel = "GDP (Lakh Crore)")
```

### Top and bottom 5 states based on % Growth

*Top 5 states on the basis of % growth*

```python
sorted_average_growth = perChange.mean().T.sort_values(ascending = False)
sorted_average_growth.head(5)
```

```python
sorted_average_growth.head(5).plot.bar(title = "% Growth of Top 5 states", ylabel = "% Growth")
```

*Buttom 5 states on the basis of % growth*

```python
sorted_average_growth.tail(5)
```

```python
sorted_average_growth.tail(5).plot.bar(title = "% Growth of Bottom 5 states", ylabel = '% Growth')
```

*States having maximum continuous growth based on mean and standard deviation*

```python
described_data = pd.DataFrame(perChange.describe())
described_data = described_data.T
described_data[(described_data['std'] < 2) & (described_data['mean'] > 7)]
```

*States having minimum continuous growth based on mean and standard deviation*

```python
described_data = pd.DataFrame(perChange.describe())
described_data = described_data.T
described_data[(described_data['std'] > 5) & (described_data['mean'] < 7)]
```
